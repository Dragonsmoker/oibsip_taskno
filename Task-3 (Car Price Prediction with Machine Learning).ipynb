```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
info = pd.read_csv('CarPrice.csv')
```

```python
plt.figure(figsize=(20,15))
sb.heatmap(info.corr(),annot=True)
plt.show()
```

```python
short_info = info[["symboling", "wheelbase", "carlength", 
             "carwidth", "carheight", "curbweight", 
             "enginesize", "boreratio", "stroke", 
             "compressionratio", "horsepower", "peakrpm", 
             "citympg", "highwaympg", "price"]]
X = np.array(short_info['price'])
Y = np.array(short_info.drop(['price'],1))
XTrain,XTest,YTrain,YTest = train_test_split(Y,X,test_size=0.21)
```

```python
model = DecisionTreeRegressor()
model.fit(XTrain,YTrain)
```

```python
model.score(XTest,model.predict(XTest))
```
